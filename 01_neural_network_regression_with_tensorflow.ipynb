# Steps in modelling with tensorflow

```python
# 1. Creating a model - define i/p and o/p layers as well as a hidden layers of a DL Model
# 2. Compiling a model - define loss function, optimizer and evaluation metric
# 3. Fitting a model - letting the model try to find patterns between X & y (features and labels)
tf.random.set_seed(42)
# 1. Create a model
model = tf.keras.Sequential([tf.keras.layers.Dense(1)])
# 2. Compile the model
model.compile(loss= tf.keras.losses.mae,optimizer = tf.keras.optimizers.SGD(),
             metrics = ['mae'])
# 3. Fit the model
model.fit(X,y,epochs=5)
```

```python
y_pred = model.predict([17])
y_pred
```

```python
# Or Alternatively
model =tf.keras.Sequential()
model.add(tf.keras.layers.Dense(1)
```

```python
#Improve the model- By altering the steps
# 1. Create a model - Increase the number of hidden units ( all called neurons), change the activation function of each layer
# 2. Compiling the model - Change the optimization function or the "learning rate" of the optimization function
# 3. Fitting the model - fit the model for more "epochs" or more data
model = tf.keras.Sequential([tf.keras.layers.Dense(1)])

model.compile(loss= tf.keras.losses.mae,optimizer = tf.keras.optimizers.SGD(),
             metrics = ['mae'])

model.fit(X,y,epochs=100)
```

```python
y_pred = model.predict([17])
y_pred
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(100, activation = 'relu'))
model.add(tf.keras.layers.Dense(1))
model.compile(loss=tf.keras.losses.mae,optimizer= tf.keras.optimizers.Adam(lr = 0.01),
             metrics = ['mae'])
model.fit(X,y,epochs=100)
```

# Evaluating a model

```python
#Visualize
#* The data, the model itself, the training of a model, the prediction of model
# Bigger data set
X = tf.range(-100,100,4)
X
```

```python
y = X+10
y
```

```python
#Visualize the data
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
###The 3 sets---- Training set(Model learns from this data 70-80%), Validation set(Model gets tunned on this i,e tweaking the model 10-15%), Test set(Model gets evaluated on this data 10-15%)
#Split the data into train and test sets
X_train = X[:40]
y_train = y[:40]
X_test = X[40:]
y_test = y[40:]
```

```python
#Visualising the data
plt.figure(figsize =(10,7))
plt.scatter(X_train, y_train, c="b",label = "Training data")

plt.scatter(X_test, y_test, c="r",label = "Testing data")
plt.legend()
```

```python
#Built a neural network
tf.random.set_seed(42)
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(10, input_shape =[1]))
model.add(tf.keras.layers.Dense(100, input_shape =[1]))
model.add(tf.keras.layers.Dense(1, activation = "relu"))
model.compile(loss= tf.keras.losses.mae, 
              optimizer = tf.keras.optimizers.Adam(lr=0.01),
              metrics=['mae'])
model.fit(X_train,y_train,epochs=100,verbose = 1)
```

```python
model.summary()
```

```python
from tensorflow.keras.utils import plot_model
plot_model(model = model, show_shapes=True)
```

```python
#Visualizing Models Prediction
y_pred = model.predict(X_test)
y_pred
```

```python
#Let's create a plotting function
def plot_predictions(train_data = X_train, train_labels = y_train, test_data = X_test, test_labels = y_test, predictions = y_pred):
    plt.figure(figsize=(10,7))
    plt.scatter(train_data, train_labels, c="b",label = "Training data")

    plt.scatter(test_data, test_labels, c="r",label = "Testing data")

    plt.scatter(test_data, predictions, c="g",label = "Predictions")
    plt.legend();
```

```python
plot_predictions()
```

```python
#Evaluating our models prediction with regression evaluation metrics
X_train, y_train
```

```python
#3 Modelling experiments:
#Model 1: 1 layer, 100 epochs
#Model 2: 2 layer, 100 epochs
#Model 3: 2 layer, 500 epochs
#Build Model 1
tf.random.set_seed(42)
model_1 = tf.keras.Sequential()
model_1.add(tf.keras.layers.Dense(1))
model_1.compile(loss = tf.keras.losses.mae,
               optimizer = tf.keras.optimizers.SGD(),
               metrics=["mae"])
model_1.fit(X_train,y_train,epochs=100)
```

```python
y_preds_1 = model_1.predict(X_test)
y_preds_1
plot_predictions(predictions = y_preds_1)
```

```python
def mae(y_true,y_pred):
    return tf.metrics.mean_absolute_error(y_true=y_test,y_pred=tf.squeeze(y_pred));
```

```python
def mse(y_true,y_preds):
    return tf.metrics.mean_squared_error(y_true=y_test,y_pred=tf.squeeze(y_pred));
```

```python
mae_1 = mae(y_test, y_preds_1)
mse_1 = mse(y_test, y_preds_1)
mae_1, mse_1
```

```python
#Build Model 2
tf.random.set_seed(42)
model_2 = tf.keras.Sequential()
model_2.add(tf.keras.layers.Dense(10))
model_2.add(tf.keras.layers.Dense(1))
model_2.compile(loss = tf.keras.losses.mae,
               optimizer = tf.keras.optimizers.SGD(),
               metrics=["mae"])
model_2.fit(X_train,y_train,epochs=100)
```

```python
y_preds_2 = model_2.predict(X_test)
y_preds_2
plot_predictions(predictions = y_preds_2)
```

```python
mae_2 = mae(y_test, y_preds_2)
mse_2 = mse(y_test, y_preds_2)
mae_2, mse_2
```

```python
#Build Model 3
tf.random.set_seed(42)
model_3 = tf.keras.Sequential()
model_3.add(tf.keras.layers.Dense(10))
model_3.add(tf.keras.layers.Dense(1))
model_3.compile(loss = tf.keras.losses.mae,
               optimizer = tf.keras.optimizers.SGD(),
               metrics=["mae"])
model_3.fit(X_train,y_train,epochs=500)
```

```python
y_preds_3 = model_3.predict(X_test)
y_preds_3
plot_predictions(predictions = y_preds_3)
```

```python
mae_3 = mae(y_test, y_preds_3)
mse_3 = mse(y_test, y_preds_3)
mae_3, mse_3
```

```python
#Compare the results
import pandas as pd
model_results = [["model_1",mae_1.numpy(),mse_1.numpy()],
                ["model_2",mae_2.numpy(),mse_2.numpy()],
                ["model_3",mae_3.numpy(),mse_3.numpy()]]
all_results = pd.DataFrame(model_results,columns=["model1","mae","mse"])
all_results
```
